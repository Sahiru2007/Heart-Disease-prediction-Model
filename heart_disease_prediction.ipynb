```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pickle
```

```python
# loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('heart.csv')
```

```python
# print first 5 rows of the dataset
heart_data.head()
```

```python
# number of rows and columns in the dataset
heart_data.shape
```

```python
# getting some info about the data
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
# statistical measures about the data
heart_data.describe()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# StandardScaler example
scaler = StandardScaler()
heart_data[['age', 'trestbps', 'chol', 'thalach', 'oldpeak']] = scaler.fit_transform(heart_data[['age', 'trestbps', 'chol', 'thalach', 'oldpeak']])
```

```python
# IQR method for 'age'
Q1 = heart_data['age'].quantile(0.25)
Q3 = heart_data['age'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
heart_data = heart_data[(heart_data['age'] >= lower_bound) & (heart_data['age'] <= upper_bound)]
```

```python
# statistical measures about the data
heart_data.describe()
```

```python
# checking the distribution of Target Variable
heart_data['target'].value_counts()
```

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
